```python
#placement data


x= np.linspace(1,10,20).reshape(-1,1)                #linspace fucn for arranging data in ascending order
y1=np.linspace(20000,800000,20)
y2=np.random.randint(10000,100000,20)
y= np.add(y1,y2).reshape(-1,1)

```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
grade=float(input('enter the grade'))
new_x=np.array([grade]).reshape(-1,1)
model.predict(new_x)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred=model.predict(x)
```

```python
plt.scatter(x,y_pred,color='red')
plt.show()
```

```python
plt.scatter(x,y)
plt.scatter(x,y_pred,color='red')
plt.show()
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color='red',linewidth=3)
plt.show()
```

```python
print("slpoe of model:-" ,model.coef_,"intercept of model:-",model.intercept_)
```
